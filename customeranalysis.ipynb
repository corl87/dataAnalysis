```python
#Kyle Dille
import numpy as np
import pandas as pd

df = pd.read_csv('/kaggle/input/customer-analytics-practice-dataset/Mall_Customers_Enhanced.csv')
boxplot = df.boxplot(column=["Estimated Savings (k$)", "Age"])

scatterPlot = df.plot.scatter(x='Annual Income (k$)', y='Estimated Savings (k$)', title='Income and Savings Correlation')

lines = df.plot.line(x='CustomerID', y='Annual Income (k$)', title='Customer ID & Annual Income Relationship')

ageGroup= df.groupby("Age Group").mean(numeric_only = True)

bar = ageGroup.plot.bar(y='Estimated Savings (k$)', title='Savings by Age Group')
genderSplit = df.groupby("Gender").mean(numeric_only=True)
pieChart = genderSplit.plot.pie(y='Annual Income (k$)', title='Annual Income Gender breakdown')

preferredCat = df.groupby("Preferred Category").mean(numeric_only=True)
piechart= preferredCat.plot.pie(y='Annual Income (k$)')
print(df)
```
