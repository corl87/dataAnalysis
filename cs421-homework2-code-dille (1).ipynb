```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

df = pd.read_csv('/kaggle/input/cs421-hw2-dataset/cs421_homework2_dataset.csv')

numColumns = df.shape[1]
numRows = df.shape[0]
print("The homework data has " + str(numColumns) + " columns and " + str(numRows) + " rows to start")

numBlanks = 0
for x in range(numRows):
    for y in range(numColumns):
        if(np.isnan(df.iloc[x, y])):
            numBlanks+=1
print("There are "+ str(numBlanks) + " blanks in our data when we start")

df = df.dropna()
numRows = df.shape[0]
print("The homework data after dropping blanks has " + str(numRows) + " rows")

df = df.drop_duplicates()
newNumRows = df.shape[0]
print("The homework data has " + str(numRows - newNumRows) + " duplicate rows")

print(str(newNumRows) +" Rows remain after dropping duplicates")
```
